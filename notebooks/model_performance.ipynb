### For entire domain

```python
# Forecast horizons
horizons = [1, 3, 6, 9, 12, 15, 18, 21, 24]
horizons_b = [1, 3, 6, 9, 12, 15, 18, 21, 24]

# Model A data
rmse_a = [0.0890, 0.2574, 0.5066, 0.7524, 0.9594, 1.1458, 1.3045, 1.4745, 1.5790]
si_a =   [0.0222, 0.0643, 0.1265, 0.1878, 0.2395, 0.2860, 0.3256, 0.3681, 0.3941]
r_a =    [0.9995, 0.9953, 0.9813, 0.9583, 0.9323, 0.9014, 0.8697, 0.8297, 0.8034]

# Model B data 
rmse_b = [0.1104, 0.2637, 0.5198, 0.7794, 1.0069, 1.1869, 1.3675, 1.5066, 1.6743]
si_b =   [0.0276, 0.0658, 0.1297, 0.1945, 0.2513, 0.2963, 0.3413, 0.3761, 0.4179]
r_b =    [0.9992, 0.9952, 0.9811, 0.9563, 0.9258, 0.8945, 0.8660, 0.8283, 0.7946]

# Plotting
fig, axes = plt.subplots(3, 1, figsize=(8, 10), sharex=True)

def plot_metric(ax, x_a, y_a, x_b, y_b, ylabel, label_letter):
    # Model A: blue
    ax.plot(x_a, y_a, color='blue', linestyle='-', linewidth=1)
    ax.scatter(x_a, y_a, color='blue', marker='o', label='Model A')

    # Model B: red
    ax.plot(x_b, y_b, color='red', linestyle='-', linewidth=1)
    ax.scatter(x_b, y_b, color='red', marker='s', label='Model B')

    ax.set_ylabel(ylabel)
    ax.set_xticks(horizons)
    ax.set_xticklabels([f'{h}h' for h in horizons], fontsize=10)
    ax.tick_params(axis='x', which='both', labelbottom=True)

    # Bold subplot label (e.g. a), b), c)) above and to the left
    ax.text(-0.08, 1.05, f'{label_letter})', transform=ax.transAxes,
            fontsize=15, va='bottom', ha='left')

    ax.legend()

# Plot with labels
plot_metric(axes[0], horizons, rmse_a, horizons_b, rmse_b, 'RMSE (m)', 'a')
axes[0].set_title('Domain-Wide Model Performance Over Forecasting Lead Times', pad=20)

plot_metric(axes[1], horizons, si_a, horizons_b, si_b, 'Scatter Index (SI)', 'b')
plot_metric(axes[2], horizons, r_a, horizons_b, r_b, 'Correlation (R)', 'c')
axes[2].set_xlabel('Forecast Hour')

plt.tight_layout()
plt.savefig("results/performance_plot/domain_wide_1dec.png", dpi=300)
plt.show()
```
